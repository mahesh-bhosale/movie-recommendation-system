```python
# Here we find the top 3 actors in every movie ie the first 3 dictionaries. And we only require the name(actual name) of actor part of the dictionaries
```

```python
def convert3(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:
            L.append(i['name'])
            counter = counter + 1
        else:
            break  
    return L        
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
# now for the crew part we only need data where job="Director"
```

```python
def fetch_director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
# now we will remove space from the words
# Eg. Sam Worthington will become SamWorthington.
# Because if there are 2 Sams it might split words and recommend you the other sam so we make them a single entity
```

```python
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []

    for i in text.split():
        y.append(ps.stem(i))

    return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
# here we need to find similarity between the two movies according to similar tag words
# The method we will use here is Vectorization 
# Every movie will represent a vector (here we will have 5000 vectors)
# vectors are basically 2D points like (x,y) coordinate
# so we will find the closest vector for a given vector for vectorization
```

```python
# Here we will use bag of words
# we will combine tags for all movies and then we calculate the frequency of the words that appears the most and extract them
# And then we take the tags of individual movie and see how many times the frequent words appears in the tags for that movie
# (Here we will find 5000 most frequent words)
# Likewise we calculate the frequency of each 5000 words in the tags for each movie
# This is called a vector (each movie is a vector) and to find recommendations, we locate the closest vectors
# We wont consider stop words here (are,and,of,to etc.) that is used for sentence formation
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
# now we will calculate the distance between two vectors in the vector space
# more the distance, less is the similarity and vice versa (cosine distance ie the angle bw them)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
simi = cosine_similarity(vectors)
```

```python
simi[0]
```

```python
# in order to keep the index intact
sorted(list(enumerate(simi[0])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] ==  movie].index[0]
    distances = simi[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]

    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```
